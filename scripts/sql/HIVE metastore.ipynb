```sql
SELECT t.TABLE_SCHEMA + '.' + t.TABLE_NAME AS TableName,
        SUM(p.ROWS) AS [RowCount]
FROM information_schema.tables t
INNER JOIN sys.partitions p ON OBJECT_ID(t.TABLE_SCHEMA + '.' + t.TABLE_NAME) = p.OBJECT_ID
WHERE t.TABLE_TYPE = 'BASE TABLE' AND p.index_id IN (0,1) -- heap or clustered index
GROUP BY t.TABLE_SCHEMA + '.' + t.TABLE_NAME
HAVING SUM(p.ROWS) > 0
ORDER BY [RowCount] DESC;
```
